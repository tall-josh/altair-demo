```python
import altair as alt
import pandas as pd
from vega_datasets import data
iris = data.iris()
```

```python
iris
```

# Facet Grid

From https://stackoverflow.com/questions/50164001/multiple-column-row-facet-wrap-in-altair

```python
alt.Chart(iris).mark_point().encode(
    x='petalLength:Q',
    y='petalWidth:Q',
    color='species:N'
).properties(
    width=180,
    height=180
).facet(
    facet='species:N',
    columns=2
)
```

Alternatively, the same chart can be specified with the facet as an encoding:

```python
alt.Chart(iris).mark_point().encode(
    x='petalLength:Q',
    y='petalWidth:Q',
    color='species:N',
    facet=alt.Facet('species:N', columns=2)
).properties(
    width=180,
    height=180,
)
```

# Combining Charts

You can combine charts horizontaly using `|` and vertically `&`. You can group with `()`

```python
cars = data.cars()
cars
```

```python
mpg = alt.Chart(cars[:50]).mark_circle().encode(
    x='Name',
    y='Miles_per_Gallon:Q',
).properties(
    width=480,
    height=220,
)

cyl = alt.Chart(cars[:50]).mark_line().encode(
    x='Displacement',
    y='Cylinders:Q',
).properties(
    width=480,
    height=220,
)

hrs = alt.Chart(cars[:50]).mark_bar().encode(
    x='Horsepower',
    y='Cylinders:Q',
).properties(
    width=480,
    height=220,
)

wei = alt.Chart(cars[:50]).mark_square().encode(
    x='Weight_in_lbs',
    y='Name',
).properties(
    width=480,
    height=220,
)
```

```python
(mpg | cyl) & (hrs | wei)
```
